```python
# 1. Load data (movie classification)
data = pd.read_csv('movie_classification.csv')
```

```python
# 2. Cek nilai duplikat
duplicate_count = data.duplicated().sum()
print(f'Jumlah nilai duplikat: {duplicate_count}')
```

```python
# 3. Menemukan null values dan buat presentase
null_counts = data.isnull().sum()
null_percentages = (null_counts / len(data)) * 100
print('Presentase nilai null:\n', null_percentages)
```

```python
null_percentage = data.isnull().mean() * 100
print("Persentase nilai kosong di setiap kolom:")
print(null_percentage)
```

```python
# 4. Drop missing value berdasarkan baris, kolom, imputasi
# Drop missing values berdasarkan baris
data_dropped_rows = data.dropna()
print(f'Jumlah baris setelah drop missing values: {len(data_dropped_rows)}')
```

```python
# Drop missing values berdasarkan kolom
data_dropped_columns = data.dropna(axis=1)
print(f'Jumlah kolom setelah drop missing values: {len(data_dropped_columns.columns)}')
```

```python
# Imputasi dengan mean, modus, median
data_imputed_mean = data.fillna(data.mean(numeric_only=True))
data_imputed_modus = data.fillna(data.mode().iloc[0])
data_imputed_median = data.fillna(data.median(numeric_only=True))
```

```python
# 5. Export data ke csv dan excel
data_dropped_rows.to_csv('movies_drop_rows_alt.csv', index=False)
data_dropped_columns.to_csv('movies_drop_columns_alt.csv', index=False)
data_imputed_mean.to_csv('movies_impute_mean_alt.csv', index=False)
data_imputed_modus.to_csv('movies_impute_mode_alt.csv', index=False)
data_imputed_median.to_csv('movies_impute_median_alt.csv', index=False)

data_dropped_rows.to_excel('movies_drop_rows_alt.xlsx', index=False)
data_dropped_columns.to_excel('movies_drop_columns_alt.xlsx', index=False)
data_imputed_mean.to_excel('movies_impute_mean_alt.xlsx', index=False)
data_imputed_modus.to_excel('movies_impute_mode_alt.xlsx', index=False)
data_imputed_median.to_excel('movies_impute_median_alt.xlsx', index=False)
```
