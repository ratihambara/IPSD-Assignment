3. Lakukan imbalance handling dengan undersampling

```python
df_imbalance = pd.DataFrame(df)
df_imbalance
```

```python
df_imbalance.sum()
```

```python
df_imbalance.describe()
```

```python
# Pisahkan data menjadi mayoritas dan minoritas
df_majority = df_imbalance[df_imbalance['Outcome'] == 0]  # Kelas mayoritas
df_minority = df_imbalance[df_imbalance['Outcome'] == 1]  # Kelas minoritas

# Melakukan undersampling pada majority class
df_majority_undersampled = resample(df_majority,
                                    replace=False,
                                    n_samples=len(df_minority),
                                    random_state=42)

# Menggabungkan kembali minority class dan undersampled majority class
df_balanced = pd.concat([df_majority_undersampled, df_minority])

# Mengecek distribusi kelas setelah undersampling
print(df_balanced['Outcome'].value_counts())
```

4. Lakukan scaling dengan robust scaler dan minmax

```python
df_scaling = pd.DataFrame(df)
df_scaling
```

```python
df_scaling.sum()
```

```python
df_scaling.describe()
```

```python
# Scaling dengan RobustScaler
scaler_robust = RobustScaler()
df_scaling_robust = scaler_robust.fit_transform(df_scaling)
df_scaling_robust = pd.DataFrame(df_scaling_robust, columns=df_scaling.columns)
print(df_scaling_robust)
```

```python
# Scaling dengan MinMaxScaler
scaler_minmax = MinMaxScaler()
df_scaling_minmax = scaler_minmax.fit_transform(df_scaling)
df_scaling_minmax = pd.DataFrame(df_scaling_minmax, columns=df_scaling.columns)
print(df_scaling_minmax)
```
