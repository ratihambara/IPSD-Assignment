```python
# Variable dan Tipe Data
x = 10
y = 3.14
z = "Data Science"

user_age = 20
pi_value = 3.14159
course_name = "Python for Data Science"

# Integers
age = 25

# Floats
height = 1.75

# Strings
name = "Ratih"

#Booleans
is_student = True

print (type(age))
print (type(height))
print (type(name))
print (type(is_student))

# Operator Aritmatika
a = 10
b = 3

# Penjumlahan
print(a + b)

# Pembagian
print(a / b)

# Pembagian Bulat
print(a // b)

# Eksponensial
print(a ** b)

# Operator Perbandingan
print(a > b) #Output: True
print(a == b) #Output: False

# Operator Logika
x = True
y = False
print(x and y)
print(x or y)

# Fungsi
# Definisi fungsi sederhana
def greet(name):
    return f"Hello, {name}!"

# Memanggil fungsi
print(greet("Ratih"))

# Pengulangan
# For Loop
fruits = ["apple", "banana", "cherry"]
for fruit in fruits:
    print(fruit)

# While Loop
counter = 0
while counter < 3:
    print(counter)
    counter += 1

# List Comprehension
numbers = [1, 2, 3, 4, 5]
double = [x * 2 for x in numbers]

# If, Elif, Else
score = 85
if score >= 90:
    print("A")
elif score >= 80:
    print("B")
else:
    print("C")

# Nested Conditionals
score = 85
attendance = 90
if score >= 80:
    if attendance >= 85:
        print("Lulus dengan nilai baik")
    else:
        print("Kehadiran kurang")
else:
    print("Tidak lulus")
```
